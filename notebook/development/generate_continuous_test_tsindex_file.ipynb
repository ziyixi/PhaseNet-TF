```python
from obspy.clients.filesystem.tsindex import Client
```

```python
client=Client("/expanse/lustre/scratch/xiziyi/temp_project/tongml_sample/timeseries.sqlite")
```

```python
client
```

```python
extents = client.get_availability_extent(network="*", channel="*")
for extent in extents:
    print("{0:<3} {1:<6} {2:<3} {3:<4} {4} {5}".format(*extent))
```

```python
from obspy import UTCDateTime
```

```python
t = UTCDateTime("2010-04-07T00:00:00")
st = client.get_waveforms("Z1", "CICA", "*", "*", t, t + 30*60)
st.plot();
```
